```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('LG-Data-Nov-22.xlsx')
df.head()
```

```python
features = ['Na', 'GLUCOSE', 'AGE_AT_LISTING', 'SEX','RACE','HEIGHT_AT_LISTING','WEIGHT_AT_LISTING',
                    'BLOOD_TYPE', 'RH_FACTOR', 'P_DIAG', 'SEC_DIAG', 'SMOKER', 'DIALYSIS_TYPE', 'CREAT', 'BILT', 
                    'AST', 'ALT', 'ALK_PHOS', 'PLATELETS','WBC','ALB','EST_GFR','MELD','HGB','INR','Final State']
newdf = df[features].dropna(axis=0, how='any')
```

```python
xPatients = newdf.loc[:, newdf.columns!='Final State']
yPatients = df['Final State']
xPatients.head()
print(len(xPatients))
```

```python
xPatients.head()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(xPatients, yPatients, random_state=0, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.model_selection import GridSearchCV

# Gaussian Classifier
#clf = RandomForestClassifier(n_estimators=100)

#clf.fit(xTrain, yTrain)
#yPredict = clf.predict(xTest)
```

```python
params_to_search={'n_estimators':[10,50,100,150],
                  'max_depth':[6,8,10,12],
                  'min_samples_split':[2,5,8,10],
                  'min_samples_leaf':[2,5,8,10],
                 }
mdlRF = RandomForestClassifier(criterion='gini',bootstrap=True)
optimized_clf = GridSearchCV(mdlRF,params_to_search,cv=5)
optimized_clf.fit(xTrain,yTrain)
```

```python
optimized_clf.best_estimator_
```

```python
score = optimized_clf.best_score_
print(score)
```

```python
optimized_clf.best_params_
```

```python
yPred = optimized_clf.predict(xTest)
#scores = cross_val_score(, x_train, y_train, cv=5)
mean_squared_error(yTest, yPred)
```
